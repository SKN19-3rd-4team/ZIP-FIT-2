```python
import json
from pathlib import Path

# JSON 파일 경로
json_file_path = Path("LH예술인특화형주택입주자모집공고문_그루하우스서대문(251119)_elements_20251204_141635.json")

# JSON 파일 읽기
with open(json_file_path, 'r', encoding='utf-8') as f:
    data = json.load(f)

# 데이터 확인
print(f"데이터 타입: {type(data)}")
print(f"데이터 키(딕셔너리인 경우): {data.keys() if isinstance(data, dict) else 'N/A'}")
print(f"데이터 길이(리스트인 경우): {len(data) if isinstance(data, list) else 'N/A'}")
print("\n첫 번째 항목 미리보기:")
if isinstance(data, list) and len(data) > 0:
    print(data[0])
elif isinstance(data, dict):
    print(list(data.items())[:3])
```
